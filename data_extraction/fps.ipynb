```python
import os, json, cv2

folder = r"D:\VN_Multi_User_Video_Search\AIC_Video"     
out_json = "fps.json"

VIDEO_EXTS = {".mp4", ".avi", ".mov", ".mkv", ".m4v", ".webm"}

fps_map = {}
for name in os.listdir(folder):
    path = os.path.join(folder, name)
    ext = os.path.splitext(name)[1].lower()
    if os.path.isfile(path) and ext in VIDEO_EXTS:
        cap = cv2.VideoCapture(path)
        fps = cap.get(cv2.CAP_PROP_FPS) or 0.0
        cap.release()
        if fps > 0:
            fps_map[os.path.splitext(name)[0]] = float(fps)

with open(out_json, "w", encoding="utf-8") as f:
    json.dump(fps_map, f, ensure_ascii=False, indent=2)

print(f"Đã lưu vào {os.path.abspath(out_json)}")
```
