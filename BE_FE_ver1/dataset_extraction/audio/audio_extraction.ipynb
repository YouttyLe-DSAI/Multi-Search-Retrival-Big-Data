```python
import os
import sys
import subprocess
from tqdm import tqdm

def convert_video_to_audio_ffmpeg(video_file, save_path, output_ext="wav"):
    """Converts video to audio directly using `ffmpeg` command
    with the help of subprocess module"""
    ffmpeg_path = r"D:\VN_Multi_User_Video_Search\ffmpeg\bin\ffmpeg.exe"  # Đường dẫn đầy đủ tới ffmpeg.exe
    subprocess.call([ffmpeg_path, "-y", "-i", video_file, f"{save_path}.{output_ext}"], 
                    stdout=subprocess.DEVNULL,
                    stderr=subprocess.STDOUT)
```

# Parse video info

```python
videos_dir = r'D:\VN_Multi_User_Video_Search\AIC_Video'
all_video_paths = dict()
for video_file in sorted(os.listdir(videos_dir)):
    if video_file.endswith('.mp4'):
        # Tách lấy phần Lxx và Vxxx từ tên file
        parts = video_file.replace('.mp4', '').split('_')
        if len(parts) == 2:
            data_part = parts[0]  # Lxx
            video_id = parts[1]   # Vxxx
            if data_part not in all_video_paths:
                all_video_paths[data_part] = dict()
            video_path_full = os.path.join(videos_dir, video_file)
            all_video_paths[data_part][video_id] = video_path_full
```

# Get audio

```python
save_dir_all = './audios'
if not os.path.exists(save_dir_all):
    os.mkdir(save_dir_all)

for key in tqdm(all_video_paths.keys()):
    save_dir = f'{save_dir_all}/{key}'

    if not os.path.exists(save_dir):
        os.mkdir(save_dir)
        
    video_paths_dict = all_video_paths[key]
    video_ids = sorted(video_paths_dict.keys())
    for video_id in tqdm(video_ids):
        video_path = video_paths_dict[video_id]
        save_path = f'{save_dir}/{video_id}.wav'
        if not os.path.exists(save_path):
            convert_video_to_audio_ffmpeg(video_path, save_path.replace('.wav', ''))
```
