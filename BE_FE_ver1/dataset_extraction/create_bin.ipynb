```python
import os
import glob
import faiss
import numpy as np
from tqdm import tqdm
```

# Clip (open_ai)

```python
feature_shape = 768
features_dir = './CLIP_features'

index = faiss.IndexFlatIP(feature_shape)

for l_folder in sorted(os.listdir(features_dir)):
    l_path = os.path.join(features_dir, l_folder)
    if not os.path.isdir(l_path):
        continue
    for feature_path in sorted(glob.glob(os.path.join(l_path, '*.npy'))):
        feats = np.load(feature_path)
        print(f"{feature_path}: shape={feats.shape}")
        if feats.size == 0:
            print(f"File rng: {feature_path}")
            continue
        index.add(feats.astype(np.float32))

faiss.write_index(index, "faiss_clip_cosine.bin")
```
